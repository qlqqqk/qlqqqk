### 제너레이터 
 이터레이터의 일종으로 일반함수의 진입점은 하나, generator는 yield를 활용해 진입점을 여러개로 만듬

```python
# iterator 
'''
iter안에 저장된 input(반복가능한 객체들 ex.str,list,tuple)에 대해서 내장함수 next로 한번씩 호출할 수 있게 만듬.
일회성으로 필요할 때 메모리 절약하며 사용가능.
'''
import sys
it = iter([0,1,2,3])
li = [0,1,2,3]
next(it),next(it),next(it),next(it)
sys.getsizeof(it),sys.getsizeof(li)
```

```python
#generator
'''
yield문 이전까지의 행동을 실행한다. 
변수에 할당할 경우 yield의 value가 저장된다.
iter 혹은 generator의 끝에 도달해 Stopiteration에 걸릴때 
next(generator,default) 디폴트값을 지정해주면 stopiteration에 걸리지않고 디폴트 값을 출력.
'''
ex = 'local'
def ex_generator():
    print('first text')
    ex = 'gen'
    yield ex 
    print('second text')
    yield 2
    print('third text')
    yield -1
    yield True
    
gen = ex_generator()
a = gen.__next__()
type(gen),next(gen),next(gen),next(gen)
```

```python
# generator local var
def ex2_gen():
    n = 0
    n+=1
    yield n
    n+=1
    yield n
    n+=1
    yield n
ex2 = ex2_gen()
next(ex2),next(ex2),next(ex2)
```

## GIL, Global Interpreter Lock

    Python의 object는 참조 횟수(reference count)를 저장하는 필드를 갖고 있다. reference count가 0가 될 때 Garbage collection으로 해당 object를 메모리에서 삭제 시킨다.  
    python interpreter를 복수의 thread가 실행하면 race condition(값이 올바르지 않게 읽히거나 쓰이는 상태), 즉 thread-safe하지 못한 상태가 발생 되기에 이러한 현상을 방지하기 위해 GIL을 사용하여, 한줄을 읽을 때마다 하나의 thread가 동작하도록 lock하는 방식을 말한다.  
    그러나 CPU의 연산 비중이 적은 외부 연산(ex. sleep)의 비중이 큰 작업을 할 때는 multithreading을 활용할 때 속도개선 효과가 있다.

### GC, Garbage Collection

    1. Reference counting
    2. Generational garbage collection
        참조 주기(reference cycle)로 인해 사용하지 않는 object의 reference counting이 0에 도달하지 않을 떄
